```python
import matplotlib.pyplot as plt
from matplotlib import cm
import pylab as pylab

import numpy as np

%matplotlib inline
```

```python
data=np.loadtxt('heatmap_serial_256.txt')
```

```python
data.shape
x=data[:,0]
y=data[:,1]
z=data[:,2]
n=int(np.sqrt(len(x)))
print(n)
z=z.reshape((n,n))
x=x.reshape((n,n))
y=y.reshape((n,n))
zmax=np.amax(z)
zmin=np.amin(z)
```

```python
fig, ax = plt.subplots()
cax = ax.imshow(z, interpolation='nearest')
cbar = fig.colorbar(cax,ticks=[zmin,0.25,0.5,0.75,zmax])
cbar.ax.set_yticklabels([zmin, '0.25','0.5','0.75', zmax])
plt.title('Serial, Size=256')
plt.imshow(z)
pylab.savefig('heatmap_serial_256.pdf')
```
